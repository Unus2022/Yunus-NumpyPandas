```python
import pandas as pd #import pandas
assults = pd.read_csv('assults.csv') #load data
```

```python
assults.head() #view first 5 rows of the dataset
```

```python
'''
1. How many regions are included in the data? The regions can be found under "Region_2013_label".
'''
assults["Region_2013_label"].nunique()
```

```python
''' 
2. List the regions in alphabetical order.
'''
sorted(assults["Region_2013_label"].unique())
```

```python
'''
3. What was the population of each region in the middle of the year 2015? Use Pandas to display
'''
region_population = assults.groupby("Region_2013_label")["Population_mid_point_2015"].sum()
region_population
```

```python
'''
4. How many crimes were committed in each region in 2015? You will find the crimes in the 
column "Victimizations_calendar_year_2015
'''
assults.groupby("Region_2013_label")["Victimisations_calendar_year_2015"].sum()
```

```python
'''
5. Show the 10 areas with the highest number of assaults. Areas are "Area_unit_2013_label".
'''
assults.groupby("Area_unit_2013_label")["Victimisations_calendar_year_2015"].sum().sort_values(ascending = False).head(10)
```

```python
'''
6. Add a column to the dataframe that shows the % number of assaults in the area compared to population in the area. 
Each row is an area.
'''
assults_copy = assults.copy()
assults_copy["percentage_assult"] = (assults_copy["Victimisations_calendar_year_2015"] / assults_copy["Population_mid_point_2015"] * 100)
```

```python
assults_copy.head()
```

```python
'''
7. Create a new dataframe that contains the number of areas (Area_unit_2013_label) that exist 
of the different types of urban area (Urban_area_2013_label).
'''
urban_area = assults.groupby("Urban_area_2013_label")["Area_unit_2013_label"].count()
df_urban_area = pd.DataFrame(urban_area)
df_urban_area = df_urban_area.reset_index()
new_dataframe = df_urban_area.rename(columns = {"Area_unit_2013_label": "Number_of_areas"})
new_dataframe
```

```python
'''
8. Find how many areas (each row is an area) had population but no raids.

'''
pop_but_no_raid = assults[(assults["Victimisations_calendar_year_2015"]==0) & (assults["Population_mid_point_2015"] > 0)]
pop_but_no_raid
```

```python
'''
Compare the total population in these with the total population in their region (Region_2013_label).
'''
total_no_raid_population = pop_but_no_raid["Population_mid_point_2015"].sum() #total population
region_data = assults.groupby("Region_2013_label")["Population_mid_point_2015"].sum()

region_df = pd.DataFrame (region_data)
region_df.reset_index(inplace = True)
region_df["total_no_raid_population"] = total_no_raid_population
region_df["region_pop_to_no_raid__total_pop"] = (total_no_raid_population/region_df["Population_mid_point_2015"])

region_df
```

```python
'''
9. Count per column how many rows there are that do not contain any data
'''
assults.isnull().sum()
```

```python
'''
10. Create a new dataframe with 10 rows. It contains the 10 rows with the
lowest number of assaults per inhabitant (Rate_per_10000_population), but more than 5000 
inhabitants
'''

pop_500_above = assults[assults["Population_mid_point_2015"] > 5000]
pop_500_abovee = pop_500_above.dropna(subset = ["Rate_per_10000_population"])
pop_500_abovee = pop_500_abovee.astype({'Rate_per_10000_population': 'int'})
final = pop_500_abovee.sort_values("Rate_per_10000_population", ignore_index = True).head(10);
final
```

```python
'''
11. Create a new dataframe "regions" that contains composite data per region (Region_2013_label). 
Add a new column containing "Rate_per_10000_population" fixed for the entire region.
'''
regions = assults.groupby("Region_2013_label").sum()
regions.reset_index(inplace=True)
regions
regions["Rate_per_10000_population"] = (regions["Victimisations_calendar_year_2015"]/regions["Population_mid_point_2015"] * 1000)
regions
```

```python
'''
12. Add a new column to the dataframe with the regions showing in % how many of all 
assaults that takes place in that region.
'''
total_assults = sum(regions["Victimisations_calendar_year_2015"])
regions["percentage_assults"] = (regions["Victimisations_calendar_year_2015"]/total_assults * 100)
regions
```
